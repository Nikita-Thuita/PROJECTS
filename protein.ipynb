Fitting a multiple linear regression
Y=0+1X1+2X2++nXn+

Where:

    Y is the dependent variable (the variable we are trying to predict).
    X1,X2,,XnX1,X2,,Xn are the independent variables (predictor variables).
    0 is the intercept (the value of Y when all independent variables are zero).
    1,2,,n1,2,,n are the coefficients (the parameters that represent the relationship between each independent variable and the dependent variable).
     is the error term (the difference between the observed value and the predicted value of YY).
