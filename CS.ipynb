Distribution of melanoma cancer incidence

```python
#plot histograms for each numeric column
x =clean_cs.columns
plt.figure(figsize=(30, 25))
for i, column in enumerate(x,1):
    i>1804 #row*column ie 164*11
    plt.subplot(5,3,i) #5 rows ,3 columns
    sns.histplot(clean_cs[column], kde=True,color="orange") #kde means kernel density estimation and allows for estimation of pdf
    plt.title(f'Histogram of {column}')
plt.tight_layout()
plt.show()
```

The length of individual bar in the histogram shows the frequency of occurrence of respective factors such as cancer deaths attributable to alcohol.Higher bars indicate higher frequency of occurrence  

plotting box plots

```python
plt.figure(figsize=(30,25))
for i, column in enumerate(clean_cs.columns, 1):
    plt.subplot(5,3,i)
    sns.boxplot(y=clean_cs[column],color='orange')
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```
